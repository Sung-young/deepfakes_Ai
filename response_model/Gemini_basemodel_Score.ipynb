```python
import pandas as pd
from sklearn.metrics import roc_curve, auc, f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# CSV  
data = pd.read_csv('/content/sample_data/output2.csv')  # 'output.csv'    

#   (Label)   (Predicted) 
y_true = data['Label']  #  
y_pred = data['Predicted']  #  

# ROC   AUC 
fpr, tpr, _ = roc_curve(y_true, y_pred)
roc_auc = auc(fpr, tpr)

# F1  
f1 = f1_score(y_true, y_pred)

#   
conf_matrix = confusion_matrix(y_true, y_pred)

#  
print(f"Gemini-1.5-Flash-002_F1 Score: {f1:.2f}")
print(f"Gemini-1.5-Flash-002_AUC: {roc_auc:.2f}")
print("Gemini-1.5-Flash-002_Confusion Matrix:")
print(conf_matrix)

#   
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
disp.plot(cmap=plt.cm.Blues)
plt.title('Gemini-1.5-Flash-002_Confusion Matrix')
plt.show()

# ROC  
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Gemini-1.5-Flash-002_Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
